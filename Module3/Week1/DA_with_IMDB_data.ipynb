**11. apply() functions**

```python
def rating_group(rating):
    if rating > 9:
        return "Excellent"
    elif rating > 8:
        return "Good"
    elif rating > 6: 
        return "Average"
    else:
        "Bad"
data['Rating_Category'] = data['Rating'].apply(rating_group)
data.head()
```

```python
data[data['Rating']>8.5].groupby('Rating_Category').size()
```
